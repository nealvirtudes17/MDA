```python
import pandas as pd

# Load Data
movie = pd.read_csv('../data/movie.csv')
# Create a Series for specific examples (Index=Title, Value=Score)
movie_series = movie.set_index('title')['imdb_score']

# -----------------------------------------------------------------------------
# PRODUCTION SUBSETTING MASTERCLASS
# -----------------------------------------------------------------------------

# 1. Advanced Boolean Logic & The "Vectorized" Between
# Standard: Use parens for bitwise logic (&, |) inside a lambda.
# Optimization: .between() is cleaner/faster than `(x >= a) & (x <= b)`
logic_subset = (movie
    .loc[lambda df_: 
        (df_['content_rating'].isin(['PG-13', 'R'])) &       # List membership
        (df_['year'].between(2010, 2016, inclusive='both')) & # Inclusive range
        (~df_['director_name'].isna())                       # Invert isna() (Not Null)
    ]
)

# 2. Column-to-Column Comparison & Arithmetic
# Standard: Compare columns directly within the lambda state (df_).
# Use Case: Finding profitable movies where Gross > Budget
col_comparison = (movie
    .dropna(subset=['gross', 'budget'])  # Pre-filter for safety
    .loc[lambda df_: 
            (df_['gross'] > df_['budget']) & 
            ((df_['gross'] / df_['budget']) > 3.0) # Column vs Column , arithmetic logic
    ,['title', 'budget', 'gross']] #column selection
)

# 3. Series Subsetting Standards
# Standard: Apply boolean masks directly to Series.
# Note: Series also support .loc for label-based or boolean filtering
series_subset = (movie_series
    .loc[lambda s: (s > 8.5) & (s.index.str.startswith('The'))] # Value filtering and index filetering
)

# 4. The Query Engine (Variables, Lists, and Math)
# Standard: Use .query() when logic involves external variables or complex "English-like" flow.
# Optimization: @variable allows injecting Python objects into the query string.
target_genres = ['Action', 'Adventure']
min_duration = 120
budget_cap = 100_000_000

query_advanced = (movie
    .query(
        "genres in @target_genres and "           # List containment
        "duration > @min_duration and "           # External variable
        "budget < @budget_cap and "
        "actor1_fb > actor2_fb" # Col-to-Col comparison
    )
    .loc[:,['title','genres','duration','budget','actor1_fb','actor2_fb']]
)

# 5. Type-Based & Regex Selection (Crucial Missing Method)
# Standard: Use select_dtypes for ML pipelines (separating numeric/categorical).
# Standard: Use filter(regex=...) for pattern matching column names.
feature_selection = (movie
    .select_dtypes(include=['number', 'category'])
    .filter(regex=r'^(movie|imdb|actor)') # Columns starting with these patterns
)
```
